```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("TMDB_ML.csv")
```

```python
df_3 = pd.read_csv("name_director.csv")
```

```python
colonnes_a_garder = ['id_imdb', 'id', 'directors', 'primaryName', "primaryProfession", "knownForTitles"]
df_2 = df_2[colonnes_a_garder]
```

```python
df_2.info()
```

```python
df_2 = df_2.dropna(subset=['id_imdb'])
```

```python
merge_df = pd.merge(df, df_2, how='left', left_on='imdb_id', right_on='id_imdb')
```

```python
merge_df.drop(columns=['id_imdb'], inplace=True)
```

```python
# On passe les dates en datetime
merge_df["release_date"] = pd.to_datetime(merge_df["release_date"])
```

```python
# On garde uniquement les films tout public
merge_df = merge_df[merge_df['adult'] == False]
```

```python
# On garde uniquement les films réalisé
merge_df = merge_df[merge_df['status'] == "Released"]
```

```python
# On garde les films après 1950
merge_df = merge_df[merge_df['release_date'] >= "1950-01-01"]
```

```python
# On garde les films qui ont plus de 20 votes
merge_df = merge_df[merge_df["vote_count"] >=15]
```

```python
# On garde les films qui ont plus de 1 en notes moyennes
# df = df[df["vote_average"] >=0]
```

```python
# On garde les films qui ont plus de 1 en popularité
merge_df = merge_df[merge_df["popularity"] >=2]
```

```python
# On garde les films qui durent plus de 60 minutes
merge_df = merge_df[merge_df["runtime"] >=60]
```

```python
# On supprime les colonnes qui ne seront pas utilisé dans le ML et le Streamlit (certains on servi de filtre en amont)
merge_df = merge_df.drop(columns = ["adult", "status", "homepage", "spoken_languages", "original_title", "production_countries"])
```

```python
merge_df = merge_df.dropna(subset=['genres'])
```

```python
merge_df = merge_df.dropna(subset=['production_companies'])
```

```python
df = pd.merge(merge_df, df_3, how='left', left_on='directors', right_on='nconst')
```
